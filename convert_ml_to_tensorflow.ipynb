```python
import pandas as pd
import numpy as np
import tensorflow as tf
import datetime
```

```python
print(tf.__version__)
```

```python
car_data = pd.read_csv('car_data.csv')
```

```python
car_data.head()
```

```python
car_data = car_data.iloc[:, 1:]
```

```python
car_data.head()
```

```python
car_data['current_year'] = datetime.datetime.today().year
```

```python
car_data.head()
```

```python
car_data['car_age'] = car_data['current_year'] - car_data['Year']
```

```python
car_data.head()
```

```python
car_data.drop(['Year'], axis=1, inplace=True)
```

```python
car_data.head()
```

```python
get_dumm_data = car_data
```

```python
get_dumm_data=pd.get_dummies(get_dumm_data, drop_first=True)
```

```python
get_dumm_data.head()
```

```python
car_data_features = get_dumm_data.iloc[:, 1:].values
```

```python
car_data_features
```

```python
car_data_labels = get_dumm_data.iloc[:, 0:1].values
```

```python
car_data_labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(car_data_features, car_data_labels, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from tensorflow.keras.layers import Input, Dense, Activation,Dropout
from tensorflow.keras.models import Model
```

```python
# Shape

input_layer = Input(shape=(car_data_features.shape[1],))
dense_layer_1 = Dense(100, activation='relu')(input_layer)
dense_layer_2 = Dense(50, activation='relu')(dense_layer_1)
dense_layer_3 = Dense(25, activation='relu')(dense_layer_2)
output = Dense(1)(dense_layer_3)
```

```python
model = Model(inputs=input_layer, outputs=output)
model.compile(loss="mean_squared_error" , optimizer="adam", metrics=["mean_squared_error"])
```

```python
history = model.fit(X_train, y_train, batch_size=2, epochs=100, verbose=1, validation_split=0.2)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
pred_train = model.predict(X_train)
```

```python
print(np.sqrt(mean_squared_error(y_train,pred_train)))
```

```python
pred = model.predict(X_test)
```

```python
print(np.sqrt(mean_squared_error(y_test,pred)))
```
